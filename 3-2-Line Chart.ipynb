```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('./Data/data-for-datavis/spotify.csv',index_col="Date",parse_dates=True)
```

<div id="bir"  style="color:blue;"> 1. Lineplot Oluşturma</div>
<ul>
<li> Verileri birbirine bağlı olacak şekilde çizgi frekansı çıkaran modüldür
<li> plt.figure ile çıktı boyutunu düzenleyebiliriz
</ul>

```python
plt.figure(figsize=(10,4))
sns.lineplot(data=data)
```

```python
plt.figure(figsize=(10,4))
plt.title("Başlık Kısmı")
plt.xlabel("Tarih")
plt.ylabel("Değer")
sns.lineplot(data=data)
```

##### plt.title("Başlık") ile çıktıya başlık eklenebilir
##### plt.xlabel("X Başlığı") ile çıktıya başlık eklenebilir
##### plt.ylabel("Y Başlığı") ile çıktıya başlık eklenebilir

```python
plt.figure(figsize=(10,4))
plt.title("Başlık Kısmı")
sns.lineplot(data=data.Despacito,label="Despacito")
sns.lineplot(data=data.Unforgettable,label="Unforgettable")
```

##### İstenilirse her data farklı farklı eklenebilir. Birden fazla lineplot kullanılabilir.
##### Label eklemek için için label etiketi kullanılabilir.
